### AnÃ¡lisis Univariado

```python
import seaborn as sns
import matplotlib.pyplot as plt

all_columns = list(df.columns)
# Suponiendo que 'df' es tu DataFrame y 'axis' es una matriz de subgrÃ¡ficas
fig, axis = plt.subplots(4, 9, figsize=(30, 10))
# Lista de variables para graficar
variables = all_columns
# Iterar sobre las variables y sus posiciones en el eje
for i, var in enumerate(variables):
    ax = axis[i // 9, i % 9]  # Determinar la posiciÃ³n en la matriz
    sns.histplot(ax=ax, data=df, x=var).set(ylabel=None)
plt.tight_layout()  # Ajustar el espaciado entre subgrÃ¡ficas
plt.show()
```
